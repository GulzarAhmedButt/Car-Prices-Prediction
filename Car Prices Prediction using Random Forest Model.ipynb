```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Car Data 2.csv')
```

```python
dataset.head(50)
```

```python
print(dataset['transmission'].unique())
print(dataset['fuel'].unique())
print(dataset['owner'].unique())
```

```python
#checking if null values are present or not
dataset.isnull().any()
```

```python
#checking unique values in categorical columns
print(dataset['name'].unique())
print(dataset['fuel'].unique())
print(dataset['seller_type'].unique())
print(dataset['transmission'].unique())
print(dataset['owner'].unique())
```

```python
dataset_copy = dataset.copy()
```

```python
dataset_copy.drop('name',axis=1,inplace=True)
```

```python
dataset_copy
```

```python
dataset_copy['fuel'].dtypes
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for col in dataset_copy.columns:
    if dataset_copy[col].dtypes == 'O':
        dataset_copy[col+'_encoded'] = le.fit_transform(dataset_copy[col])
```

```python
dataset_copy.drop(dataset_copy.select_dtypes('O'),axis=1,inplace=True)
```

```python
dataset_copy[dataset_copy['owner_encoded']==5]
```

```python
sns.heatmap(dataset_copy.corr(),annot=True)
```

```python
Y=dataset_copy.iloc[:,1].values
dataset_copy.drop('selling_price',axis=1,inplace=True)
X = dataset_copy.iloc[:,:6].values
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100,random_state = 0)
regressor.fit(X_train,Y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
accuracy = regressor.score(X_test,Y_test)
```

```python
accuracy
```

```python
plt.scatter(X_train[:,0],Y_train,color = 'red')
plt.plot(X_train[:,0],regressor.predict(X_train),color = 'blue')
```

```python
from sklearn.feature_selection import SelectKBest,f_regression
bestfeatures = SelectKBest(score_func=f_regression,k=6)
fit = bestfeatures.fit(X_train,Y_train)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(dataset_copy.columns)
```

```python
features = pd.concat([dfcolumns,dfscores],axis=1)
features.columns = ['Spec','Score']
```

```python
print(features.nlargest(5,'Score'))
```
